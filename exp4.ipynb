```python
#Problem 1
#A college event has a list of 10 students who registered multiple by mistake write a program using set to store participants name and display only unique participants
registered_students = [
    "Alice ", "Bob ", "Charlie", "David ", "Eve ",
    " Smith", " White", " Miller", "Johnson", "Green"
]

print(f"Original list of registered students (with potential duplicates):\n{registered_students}")
print(f"Total registrations: {(registered_students)}\n")

unique_participants = set(registered_students)

print(f"Unique participants for the event (after removing duplicates):\n{unique_participants}")
print(f"Total unique participants: {(unique_participants)}\n")

list_of_unique_participants = sorted(list(unique_participants))
print(f"Unique participants as a sorted list:\n{list_of_unique_participants}")
```

```python
#Problem 2
#three students choose elcectibe subjects find the common subjects chosen by all studentws using set operation concept used: set intersection (&)'''
student1 = {"Mathematics", "Physics", "Chemistry", "Computer Science"}

student2 = {"Physics", "Biology", "Computer Science", "English"}

student3 = {"Chemistry", "Computer Science", "Mathematics", "History"}

print(f"Student 1: {student1}")
print(f"Student 2: {student2}")
print(f"Student 3: {student3}\n")

common_subjects = student1 & student2 & student3

print(f"Common subjects chosen by all three students: {common_subjects}")
```

```python
#problem 3
#one set contains students in the cricket club another in the football club find students who are :1 in both clubs 2. in only one club concept used: intersection,symmetric difference cricket={"amit","rahul","sneha)
cricket_club = {"amit", "rahul", "sneha"}
football_club = {"rahul", "karan", "pooja"}

print(f"Cricket Club Members: {cricket_club}")
print(f"Football Club Members: {football_club}\n")

# 1. Students in both clubs (Intersection)
students_in_both = cricket_club & football_club
print(f"Students in both clubs (intersection): {students_in_both}")

# 2. Students in only one club (Symmetric Difference)
students_in_only_one = cricket_club ^ football_club
print(f"Students in only one club (symmetric difference): {students_in_only_one}")
```

```python
''' Problem Statement 4: Absent Students
Given a set of all students and a set of present
students, find the absent students'''
all_students = {"Alice", "Bob", "Charlie", "David", "Eve"}
present_students = {"Bob", "David"}

print(f"All Students: {all_students}")
print(f"Present Students: {present_students}\n")

absent_students = all_students - present_students

print(f"Absent Students: {absent_students}")

```

```python
''' Problem 5: Remove the invalid entries.
A set contains valid course codes
Remove a discontinued course code from the set.
Concept used: remove()/discard()'''
valid_course_codes = {"CS101", "CS102", "CS103"}
discontinued_course_code = "CS102"

print(f"Valid Course Codes: {valid_course_codes}\n")
print(f"Discontinued Course Code: {discontinued_course_code}\n")

valid_course_codes.remove(discontinued_course_code)

print(f"Updated Valid Course Codes: {valid_course_codes}")

```
